## Simulation scheme for genome dating project

We simulate two tree topologies that are as different as possible, but which have two nodes (plus the root) in common. These shared nodes are important to compare the estimates among analyses.

One way to do this is by specifying a data set of 50 taxa in beast, constraining monophyly of two nodes and sampling from the prior. From these samples we select the two trees that are most different. 

We simulate sequence data along the two trees. The sequences should evolve under different rates or sequence lengths, such that they have different levels of information content. This will result in esimtates of tree topology using individual genes, that will not exactly match the true tree, so we do not need to simulate the 'genes' under different topologies. The motivation behind this method is that there are sets of genes that have the same 'true topology' (e.g. mitochondrial vs. nulear genes with out ILS).

The ages of the two fixed nodes fixed in BEAST. This guarantes that our comparisson of the ages of the two nodes are valid. We also need to fix the age of the root, which we will use as a calibration. The simulations using BEAST also have the advantage that the chronogram used for simulation is always ultrametric, which is not the case for trees after NNI moves.

For example, I ran something similar in BEAST and selected two trees at random. The age of node 1 is 10, that of node 2 is 20, and that for the root is 100:

```r
library(NELSI)
sim_trees <- read.nexus('s1.trees')
tr1 <- sim_trees[[50]]
tr2 <- sim_trees[[100]]
dist.topo(tr1, tr2)
```

The trees have a large tree topology distance, but they share two nodes; one that contains taxa t1 to t10, and one containing t1 to t39. This way there are two nodes with the same height that exist in all trees:

```r
plot(tr1, cex = 0.7)
tips_node_1 <- getMRCA(tr1, tr1$tip.label[which(tr1$tip.label %in% paste0('t', 1:10))])
tips_node_2 <- getMRCA(tr1, tr1$tip.label[which(tr1$tip.label %in% paste0('t', 1:39))])

nodelabels(node = tips_node_1, round(allnode.times(tr1)[tips_node_1], 1))
nodelabels(node = tips_node_2, round(allnode.times(tr1)[tips_node_2], 1))
nodelabels(node = min(tr1$edge[, 1]), round(max(allnode.times(tr1)), 1))
```

```r
plot(tr2, cex = 0.7)
tips_node_1 <- getMRCA(tr2, tr2$tip.label[which(tr2$tip.label %in% paste0('t', 1:10))])
tips_node_2 <- getMRCA(tr2, tr2$tip.label[which(tr2$tip.label %in% paste0('t', 1:39))])

nodelabels(node = tips_node_1, round(allnode.times(tr2)[tips_node_1], 1))
nodelabels(node = tips_node_2, round(allnode.times(tr2)[tips_node_2], 1))
nodelabels(node = min(tr2$edge[, 1]), round(max(allnode.times(tr2)), 1))
```

Now we have two topology clusters. We can simulate three pacemakers within each, such that we have six simulation phylograms. For each phylogram we can generate about 100 'genes' (or 20 if this is too many) and esitmate the topology for each gene individually. As a result we have 600 estimated topologies. We then do the following:

- Calculate distance between all topologies and identify the clusters. There should be two.

    - For each cluster find the pacemakers. This involves concatenating the genes to estimate a topology, and optimising branch lengths for each gene on this topology.
    
    - Analyse each topology cluster:
        
        - ignoring pacemakers
        
        - specifying pacemakers by randomly assigning genes to three artificial pacemakers
        
        - specifying pacemakers correctly (according to clockstar)
        
    - We also analyse the concatenated data by ignoring topology clusters:
    
        - ignore pacemakers
        
        - assign genes to three artificial pacemakers
        
        - use clockstar to identify pacemakers and use in analysis

For all of these analyses we calibrate the root, or node 2, and estimate the ages of the remaining two nodes. All other nodes will probably not exist in most analyses. A potential problem is that some nodes might not exist exactly, as specified here, such as node 1. In that case, we take the tmrca of the corresponding taxa, even if this includes some taxa that we did not specify in our simulations.
